```python
import google.generativeai as genai
import os

genai.configure(api_key=os.getenv("GOOGLE_API_KEY_1"))

generation_config = {
    "temperature": 0.6,
    "top_p": 0.9,
    "max_output_tokens": 65536,
}
# safety_settings = [
#     {"category": "HARM_CATEGORY_HARASSMENT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
#     # add more as needed...
# ]

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash-preview-05-20",
    generation_config=generation_config,
    # safety_settings=safety_settings
)
```

### Creation of Needles

```python
prompt = """Write a collection of {collection_size} facts about VVIPs of {country} on {topic}, 
where each group of {num_related_facts} facts are related. Write a single/multi-hip factual reasoning
question and answer (one or 2 word answer) for each group of related facts. 
The answer should not be explicitly mentioned among the facts. The answer needs to be derived through reasoning. 
Enclose each group of related facts, question, answer, and reasoning
inside a python dictionary (e.g., {{'facts': [list of facts], 'question': [question], 'answer': [answer], 'reasoning': [reasoning]}}),
and return a list of dictionaries."""

collection_size = 30
country_list = ["USA", "China", "Germany", "India", "Japan", "UAE"]
num_related_facts = 3
topic_list = ["politics", "business", "sports"]

for country in country_list:
    for topic in topic_list:
        formatted_prompt = prompt.format(collection_size=collection_size, country=country, topic=topic, num_related_facts=num_related_facts)
        print(formatted_prompt)
    #     break
    # break
        response = model.generate_content(formatted_prompt)
        print(response.text)
        break
    break
```
